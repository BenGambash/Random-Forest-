```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
df = pd.read_csv('USData.csv')
```

```python
df = df.drop('education', axis = 1)
df = df.drop('fnlwgt', axis = 1)
df = df.drop('CustomerID', axis = 1)
df = df.dropna()
df.head()
```

```python
df
```

```python
d = {
    "Private":8.0,
    "Self-emp-not-inc":7.0,
    "Self-emp-inc":6.0,
    "Federal-gov":5.0,
    "Local-gov":4.0,
    "State-gov":3.0,
    "Without-pay":2.0,
    "Never-worked":1.0
    }
df["workclass"]=df["workclass"].map(d)
df.head()
```

```python
df = pd.get_dummies(df,columns=['marital-status','occupation','relationship','race','sex','native-country'])
df.head()
```

```python
d1 = {'<=50K': 0,'>50K': 1.0}
df['income'] = df['income'].map(d1)
df.head()
```

```python
y = df['income']
x = df.drop('income',axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split( x, y, test_size = 0.3, random_state = 204440705)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
criterion = ['gini','entropy']
n_est = [50,100,400,700,1000]
depth = [2,4,10,12,16] 
```

```python
T50 = []
T100 = []
T400 = []
T700 = []
T1000 = []
R50=[]
R100=[]
R400=[]
R700=[]
R1000=[]
for i in criterion:
    print('criterion:',i)
    for j in n_est:
        print('n_estimate:',j)
        for n in depth:
            print('depth',n)
            T = RandomForestClassifier(criterion=i,n_estimators=j,max_depth=n,oob_score=True,random_state=204440705)
            T.fit(X_train,y_train)
            if i == 'gini':
                if j == 50 :
                    T50.append(1-T.oob_score_)
                elif j == 100:
                    T100.append(1-T.oob_score_)
                elif j == 400:
                    T400.append(1-T.oob_score_)
                elif j == 700:
                    T700.append(1-T.oob_score_)
                else:
                    T1000.append(1-T.oob_score_)
            else:
                if j == 50 :
                    R50.append(1-T.oob_score_)
                elif j == 100:
                    R100.append(1-T.oob_score_)
                elif j == 400:
                    R400.append(1-T.oob_score_)
                elif j == 700:
                    R700.append(1-T.oob_score_)
                else:
                    R1000.append(1-T.oob_score_)
        print()
```

```python
resultsgini = {
    'depth': depth,
    'T50': T50,
    'T100': T100,
    'T400': T400,
    'T700': T700,
    'T1000': T1000
}
resultsentropy = {
    'depth': depth,
    'R50': R50,
    'R100': R100,
    'R400': R400,
    'R700': R700,
    'R1000': R1000
}
```

```python
df1 = pd.DataFrame(resultsgini,columns = ['depth','T50','T100','T400','T700','T1000'])
df2 = pd.DataFrame(resultsentropy,columns = ['depth','R50','R100','R400','R700','R1000'])
```

```python
print('The oob_score_error by gini')
df1
```

```python
minoobgini = ''
for i in df1:
    for j in range(len(df1[i])):
        if df1.min().min() == df1[i][j]:
            n_est_gini = int(i[1:])
            depth_gini = int(df1['depth'][j])
            oob_score_error_gini = df1[i][j]
            minoobgini=df1[i][j]
print('gini')
print('n_est',n_est_gini)
print('depth',depth_gini)
print('the_min_oob_score_error',oob_score_error_gini)
```

```python
print('The oob_score_error by entropy')
df2
```

```python
minoobentropy = ''
for i in df2:
    for j in range(len(df2[i])):
        if df2.min().min() == df2[i][j]:
            n_est_entropy = int(i[1:])
            depth_entropy = int(df2['depth'][j])
            oob_score_error_entropy = df2[i][j]
            minoobentropy = df2[i][j]
print('entropy:')
print('n_est',n_est_entropy)
print('depth',depth_entropy)
print('the_min_oob_score_error',oob_score_error_entropy)
```

```python
if minoobentropy > minoobgini:
    print('the best result with criterion gini :',minoobgini)
    print('n_est',n_est_gini)
    print('depth',depth_gini)
    print('the_oob_score_error',oob_score_error_gini)
    criterionc = 'gini'
    n = n_est_gini
    dp = depth_gini
else:
    print('the best result with criterion entropy :',minoobentropy)
    print('n_est',n_est_entropy)
    print('depth',depth_entropy)
    print('the_oob_score_error',oob_score_error_entropy)
    critetionc = 'entropy'
    n = n_est_entropy
    dp = depth_entropy
    
```

```python
#ככל שיהיו יותר עצים המודל ידע לחלק יותר טוב את הדאטה ובנוסף ככל שיהיו יותר עומקים הוא יוכל להביא תוצאה מדויקת. החסרון שאני יכול להיות אוברפיטד וברגע שנרצה לבחון את המודל על נתונים חדשים אנו מסכנים בניבוי פחות טוב מהסיבה שהוא למד יותר מדי טוב את הנתונים שנתנו בפעם הראשונה. הדרך שלנו לבחון את המודל היא לתת לו נתונים חדשים לגמרי שאנחנו יודעים את התוצאה שלהם על מנת שנוכל לבחון אותו 
```

```python
from sklearn.metrics import confusion_matrix
clf = RandomForestClassifier(n_estimators=n,max_depth=dp,oob_score=True,random_state=204440705,criterion=critetionc)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
confusion_matrix(y_test,y_pred)
```
